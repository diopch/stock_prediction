```python
from stock_prediction.data_prep_api import Data_Prep_Api
from stock_prediction.tradding_app import best_stocks, true_returns, portfolio
from stock_prediction.workflow import data_collection, call_arima, arima_to_app, run_all
from stock_prediction.tradding_app import true_returns, portfolio, best_stocks
import yfinance as yf
from datetime import datetime 
import pandas as pd 
import numpy as np
```

```python
# this notebook show you how works the data_collection function 
# the date is the first date we want for all the TS
# period is the nb of days to compute the stats
```

```python
#load the data 
dict_hard_data, dict_prep_data, df_es50 = data_collection('2021-05-31', 20)
```

```python
# run the arima model
arima_df = call_arima('2021-05-31', dict_prep_data, alpha=0.05)
```

```python
arima_df
```

```python
# organize arima result per day and per stocks for the predictive returns (our simulation)
final_pred = arima_to_app('2021-05-31', '2021-06-11', arima_df, true = False)
```

```python
# organize arima result per day and per stocks for the true returns 
#(the best portfolio that we would have to choose to make the most)
# only BUY (against our strategy BUY & SELL)
final_true = arima_to_app('2021-05-31', '2021-06-11', arima_df, true = True)
```

```python
# retrieve all the data needed 
df_close_prices, df_true_returns = true_returns('2021-05-31', '2021-06-11', dict_hard_data)
```

```python
# construct our portfolio (best 10 predicted returns in absolute / BUY and SELL for one day)
best_pred = best_stocks(final_pred, sell=True, eq_weight=False)
```

```python
# construct the best portfolio for the day but only BUY (if market down, must BUY anyway)
best_true = best_stocks(final_true, sell=False, eq_weight=True)
```

```python
# our simulation of trading 
portfolio_pred = portfolio(df_close_prices, df_true_returns, best_true, best_pred, 10000, true=False)
```

```python
# the simulation of trading for the best portfolio 
portfolio_true = portfolio(df_close_prices, df_true_returns, best_true, best_pred, 10000, true=True)
```
